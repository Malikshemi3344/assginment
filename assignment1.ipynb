```python
# import libraries
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt

# load data from CSV file into a dataframe
data = pd.read_csv('hmza.csv')

marriage = data['Marriage']
divorce = data['Divorce']
unmarriad = data['Unmarriad']
year = data['Year']

# create a bar plot and give axis, color and width of each line 

fig = plt.figure()
pyplot = fig.add_axes([0.1,0.1,0.9,0.9])
pyplot.plot(year,marriage,'r', lw = 2,label = 'Marriage')
pyplot.plot(year,divorce,'b', lw = 2,label = 'Divorce')
pyplot.plot(year,unmarriad,'g', lw = 2, linestyle='--',label = 'Unmarriad')

# Set the axis labels and title

pyplot.set_xlabel('YEARS')
pyplot.set_ylabel('COUNT')
pyplot.set_title('MARRIAGE,DIVORCE & UNMARRIAD RATIO')

# add a legend

pyplot.legend()

# Show the plot

plt.show()
```

```python

#import libraries 
import pandas as pd
import matplotlib.pyplot as plt

# load data from CSV file into a dataframe 
mydf = pd.read_csv('csv.csv')
#myfreq is dataframe that store date and value of column name 
myfreq = mydf['Name'].value_counts()
#that give percentage and what type of chart 

myfreq.plot(kind = 'pie', ylabel = "", autopct = '%1.1f%%')
#title of chart 
plt.title('plastic making')
#legend that give location and size of box
plt.legend(loc = 'upper left', bbox_to_anchor = (1.2, 1))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file into a pandas dataframe
df = pd.read_csv('bbbb1.csv')

# Pivot the dataframe to create a matrix with Sales Category as columns, Year and Country as rows, and Sales Value as values
df_pivot = df.pivot(index = ['Time', 'Country'], columns='Sales Category', values='Sales Value')

# Create the stacked bar plot
ax = df_pivot.plot(kind = 'bar', stacked = False)

# Set the axis labels and title
ax.set_xlabel('Year, Country')
ax.set_ylabel('Sales Value')
ax.set_title('Sales by Category and Country')

#add legend

plt.legend()

# Show the plot

plt.show()
```
